```python
import pandas as pd
import numpy as np

df = pd.read_csv("../music_genre_dataset.csv", sep=',')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (10,7))
sns.histplot(df['BPM'], bins = 5, kde = True, color = 'blue',)
plt.title("Гистограмма BPM")
plt.show()
```

```python
plt.figure(figsize = (10,8))
plt.plot(df['RMS Energy'],df['Zero-Crossing Rate'])
plt.title('Нелинейный график')
plt.xlabel('RMS Energy')
plt.ylabel("Zero-Crossing Rate")
plt.show()
```

```python
plt.figure(figsize = (15,8))
sns.boxplot(x = 'BPM', y = 'RMS Energy', data = df)
plt.title('Боксплот')
plt.xticks(rotation=90)
plt.show()
```
